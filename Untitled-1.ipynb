```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
df = pd.read_csv("Churn_Modelling.csv")
```

```python
# preview first rows
df.head()

# dataset shape
df.shape

# column names
df.columns

# data types
df.info()

# summary statistics
df.describe()
```

```python
# missing values
df.isnull().sum()

# duplicate rows
df.duplicated().sum()
```

```python
df = df.drop(["RowNumber", "CustomerId", "Surname"], axis=1)
```

```python
#overall churn

df['Exited'].value_counts().plot(kind='bar')
plt.title("Customer Churn Count")
plt.xlabel("Exited (1 = churned)")
plt.ylabel("Number of Customers")
plt.show()
```

```python
#age of churned customers

plt.hist(df[df['Exited']==1]['Age'], bins=20)
plt.title("Age Distribution of Churned Customers")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
#balance vs churn

plt.scatter(df['Balance'], df['Exited'])
plt.title("Balance vs Churn")
plt.xlabel("Balance")
plt.ylabel("Exited")
plt.show()
```

```python
#products vs churn rate

df.groupby('NumOfProducts')['Exited'].mean().plot(kind='bar')
plt.title("Churn Rate by Number of Products")
plt.ylabel("Churn Rate")
plt.show()
```
